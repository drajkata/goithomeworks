13. Utwórz nowy DataFrame, w którym dla danych pogrupowanych według kolumny "Stanowisko" wykonaj agregację danych za pomocą metody [agg](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html) i znajdź minimalne i maksymalne wartości w kolumnie "Wynagrodzenie.za.miesiąc"

```python
polling_group_by_role = polling.groupby(by="Role")
polling_group_by_role_salary_min = polling_group_by_role["Salary/month"].agg("min").reset_index().rename(columns={"Salary/month" : "Salary/month: min"})
polling_group_by_role_salary_max = polling_group_by_role["Salary/month"].agg("max").reset_index().rename(columns={"Salary/month" : "Salary/month: max"})
polling_group_by_role_salary_min_max = pd.concat([polling_group_by_role_salary_min, polling_group_by_role_salary_max])
print(polling_group_by_role_salary_min_max)
```

14. Utwórz funkcję fill_avg_salary, która zwróci średnią miesięczną pensję. Użyj jej w metodzie [apply](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html) i utwórz nową kolumnę "avg".

```python
polling_group_by_role = polling.groupby(by="Role")
polling_group_by_role_salary_min = polling_group_by_role["Salary/month"].agg("min").reset_index().rename(columns={"Salary/month" : "Salary/month: min"})
polling_group_by_role_salary_max = polling_group_by_role["Salary/month"].agg("max").reset_index().rename(columns={"Salary/month" : "Salary/month: max"})
polling_group_by_role_salary_mean = polling_group_by_role["Salary/month"].apply(np.mean).reset_index().rename(columns={"Salary/month" : "avg"})
polling_group_by_role_salary = pd.concat([polling_group_by_role_salary_min, polling_group_by_role_salary_max, polling_group_by_role_salary_mean])
print(polling_group_by_role_salary)
```

## Dodatkowe wykresy

```python
polling_group_by_role = polling.groupby(by="Role")
df = polling_group_by_role["Salary/month"].agg("mean").reset_index().rename(columns={"Salary/month" : "Salary/month: mean"})
df_mean = df["Salary/month: mean"]
df_role = df["Role"]

plt.barh(
    df_role,
    df_mean,
    height=0.5,
    linewidth=0.5,
    edgecolor='black',
)
plt.xlabel("Średnia", fontsize="large", color="midnightblue")
plt.ylabel("Rola", fontsize="large", color="midnightblue")
plt.title("Średnia wynagrodzeń wg poszczególnych zawodów", fontsize=15)
plt.show()
```

```python
polling_group_by_role = polling.groupby(by="Role")
df = polling_group_by_role["Salary/month"].agg("max").reset_index().rename(columns={"Salary/month" : "Salary/month: max"}).nlargest(10, "Salary/month: max")
df_max = df["Salary/month: max"]
df_role = df["Role"]

plt.bar(
    df_role,
    df_max,
)
plt.xlabel("Kwota", fontsize="large", color="blue")
plt.ylabel("Rola", fontsize="large", color="blue")
plt.title("Top 10 zawodów z najwyższymi zarobkami na miesiąc", fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

```python
count_role = polling.groupby(by="Role").size().nlargest(10)
print(count_role)

labels = count_role.index
data = count_role.values
plt.plot(
    labels,
    data,
    linestyle=":",
    color="#562558",
    marker="o"
)
plt.xticks(rotation=90)
plt.title("Top 10 zawodów biorących udział w ankiecie")
plt.xlabel("Rola")
plt.ylabel("Wartość")
plt.show()
```

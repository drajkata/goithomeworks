```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

excel_file = "Fertility_rate.xlsx"
df = pd.read_excel(excel_file, names=["Regions", "2011", "2012", "2013", "2015", "2017", "2019", "2021"])
```

1. Pierwsze wiersze tabeli:

```python
print(df.head())
```

2. Liczba kolumn i wierszy:

```python
print(df.shape)
```

3. Zastąpienie w tabeli wartościi "_" wartością NaN:

```python
df = df.replace("_", np.nan)
print(df)
```

4. Typy wszystkich kolumn:

```python
print(df.dtypes)
```

5. Zamiana kolumn nienumerycznych na numeryczne:

```python
df["2011"] = pd.to_numeric(df["2011"], errors='coerce')
df["2012"] = pd.to_numeric(df["2012"], errors='coerce')
df["2013"] = pd.to_numeric(df["2013"], errors='coerce')
df["2015"] = pd.to_numeric(df["2015"], errors='coerce')
df["2017"] = pd.to_numeric(df["2017"], errors='coerce')
df["2019"] = pd.to_numeric(df["2019"], errors='coerce')
df["2021"] = pd.to_numeric(df["2021"], errors='coerce')
print(df)
print(df.dtypes)
```

6. Proporcja pustych miejsc w każdej kolumnie:

```python
df_null = pd.isnull(df)
df_rows = df.shape[0]

print("Dla 2011: ", round(df_null["2011"].sum()/df_rows, 4))
print("Dla 2012: ", round(df_null["2012"].sum()/df_rows, 4))
print("Dla 2013: ", round(df_null["2013"].sum()/df_rows, 4))
print("Dla 2015: ", round(df_null["2015"].sum()/df_rows, 4))
print("Dla 2017: ", round(df_null["2017"].sum()/df_rows, 4))
print("Dla 2019: ", round(df_null["2019"].sum()/df_rows, 4))
print("Dla 2021: ", round(df_null["2021"].sum()/df_rows, 4))
```

7. Usuń z tabeli dane dla całego kraju, ostatni wiersz tabeli - brak takiej informacji w tabeli.

8. Zastąpienie brakujących danych w kolumnach średnimi wartościami tych kolumn.

```python
df_fill_mean = df.fillna({"2011": df["2011"].mean(), "2015": df["2015"].mean(), "2017": df["2017"].mean(), "2019": df["2019"].mean(), "2021": df["2021"].mean()})
print(df_fill_mean)
```

9. Lista regionów, w których wskaźnik urodzeń w  2019 był wyższy niż średnia dla Ukrainy.

```python
fertility_rate_ukraine_mean = round(df_fill_mean["2019"].mean(),4)
print("Średnia urodzeń dla Ukrainy w 2019r. to: ", fertility_rate_ukraine_mean)
condition = (df["2019"] > fertility_rate_ukraine_mean)
regions = df.loc[condition, ["Regions", "2019"]]
print("Lista regionów ze wskaźnikiem powyżej średniej:\n", regions)

import matplotlib.pyplot as plt
plt.plot(
    regions["Regions"],
    regions["2019"],
    marker=".",
    linestyle="None",
)

plt.xlabel("Wskaźnik urodzeń", fontsize="small", color="midnightblue")
plt.ylabel("Region", fontsize="small", color="midnightblue")
plt.title("Regiony, dla których wskaźnik urodzeń w 2019r. jest wyższy niż śrenia Ukrainy w tym roku", fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

10. Region z najwyższym wskaźnikiem urodzeń w 2015r.

```python
condition = df["2015"].idxmax()
print("Region: ", df.loc[condition, "Regions"])
```

11. Wykres słupkowy wskaźnika urodzeń wg regionów w 2019r.

```python
plt.barh(
    df["Regions"],
    df["2019"],
)

plt.xlabel("Wskaźnik urodzeń", fontsize="small", color="midnightblue")
plt.ylabel("Region", fontsize="small", color="midnightblue")
plt.title("Wskaźnik urodzeń wg regionów Ukrainy w 2019r.", fontsize=15)
plt.show()
```
